```python
!pip install transformers
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python
# Load the pre-trained GPT-2 model and tokenizer
model_name = "gpt2"  # You can use other models as well, such as GPT-3 or GPT-Neo
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
# Initialize the text-generation pipeline
text_generator = pipeline("text-generation", model=model, tokenizer=tokenizer, device=-1)
```

```python
# List of input text prompts
texts = [
    "Once upon a time,",
    "The future of AI is",
    "In a distant land, there lived a",
    "It was a dark and stormy night,"
]
```

```python
# Generate text based on the input prompts
results = text_generator(texts, max_length=50, num_return_sequences=1)
```

```python
# Display the generated text for each input
for text, result in zip(texts, results):
    print(f"Input Text: {text}\nGenerated Text: {result[0]['generated_text']}\n")
```
